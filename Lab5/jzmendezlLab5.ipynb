# NVIDIA

```python
class BootCampTask(QCAlgorithm):

    def Initialize(self):
        self.SetStartDate(2019, 1, 1)
        self.SetEndDate(2021, 1, 1)

        #Select NVDA minute resolution data and set it to Raw normalization mode
        self.nvda = self.AddEquity("NVDA", Resolution.Minute)
        self.nvda.SetDataNormalizationMode(DataNormalizationMode.Raw)
        

    def OnData(self, data):
        
        #Place an order for 100 shares of NVDA and print the average fill price
        if not self.Portfolio.Invested:
            self.MarketOrder("NVDA", 100)
            
            #Debug the AveragePrice of NVDA
            self.Debug(str(self.Portfolio["NVDA"].AveragePrice))
            
```

# Portafolio 2 aos

```python
class BootCampTask(QCAlgorithm):

    def Initialize(self):
        self.SetStartDate(2019, 1, 1)  # Set Start Date
        self.SetEndDate(2121, 1, 1) #Set End Date
        self.SetCash(1000000)  # Set Strategy Cash
        
        
        self.nflx = self.AddEquity("NVDA", Resolution.Minute)
        self.nflx.SetDataNormalizationMode(DataNormalizationMode.Raw)
        
        self.tsla = self.AddEquity("TSLA", Resolution.Minute)
        self.tsla.SetDataNormalizationMode(DataNormalizationMode.Raw)
        
        self.amzn = self.AddEquity("DIS", Resolution.Minute)
        self.amzn.SetDataNormalizationMode(DataNormalizationMode.Raw)
        
        self.goog = self.AddEquity("AMZN", Resolution.Minute)
        self.goog.SetDataNormalizationMode(DataNormalizationMode.Raw)
        
        self.aapl = self.AddEquity("GOOG", Resolution.Minute)
        self.aapl.SetDataNormalizationMode(DataNormalizationMode.Raw)
        
    def OnData(self, data):

        if not self.Portfolio.Invested:
            self.SetHoldings("NVDA", 0.2)
            self.SetHoldings("TSLA", 0.2)
            self.SetHoldings("DIS", 0.2)
            self.SetHoldings("AMZN", 0.2)
            self.SetHoldings("GOOG", 0.2)
            
            
            self.Log("Active Pos NVIDIA " + str(self.Portfolio["NVDA"].Quantity));
            self.Log("Active Pos TESLA " + str(self.Portfolio["TSLA"].Quantity));
            self.Log("Active Pos DISNEY " + str(self.Portfolio["DIS"].Quantity));
            self.Log("Active Pos AMAZON " + str(self.Portfolio["AMZN"].Quantity));
            self.Log("Active Pos GOOGLE " + str(self.Portfolio["GOOG"].Quantity));
```
